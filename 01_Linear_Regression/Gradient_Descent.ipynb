```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# setting the seaborn style
sns.set_theme(style="darkgrid")
```

```python
x = np.arange(10)
print(x)
```

```python
# example: y = f(x) = (x-5)^2 + 3 = X^2-10X+28
```

```python
X = np.arange(10)
Y = (x-5)**2 + 3

print(X, Y)
```

```python
 # Visualize 
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Gradient Descent > Here when X = 5, Y will be min.
# init x with any random value
# learning rate (n) = 0.1
```

```python
x = 0
lr = 0.1

plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x-lr*grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)

plt.show()
```

```python
x = 9   # change the x value to 9
lr = 0.1

plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x-lr*grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)

plt.show()
```

```python
# change the learning rate to 0.5, then the Convergence will be fast

x = 0
lr = 0.5

plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x-lr*grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)

plt.show()
```

```python
# increase the learning rate to a much higher number 
# change the learning rate to 0.5, then the Convergence will be fast

x = 0
lr = 4

plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x-lr*grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)

plt.show()
```

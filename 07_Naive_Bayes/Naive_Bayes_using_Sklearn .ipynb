```python
import pandas as pd
import numpy as np
```

```python
golf = pd.read_csv("golf.csv")
golf
```

```python
# Sklearn cannot work with text data and needs conversion to numeric, hence we have 'lable encoder' in Sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
golf['Outlook'] = le1.fit_transform(golf['Outlook'])
```

```python
le2 = LabelEncoder()
golf['Temperature'] = le2.fit_transform(golf['Temperature'])
```

```python
le3 = LabelEncoder()
golf['Humidity'] = le3.fit_transform(golf['Humidity'])
```

```python
le4 = LabelEncoder()
golf['Windy'] = le4.fit_transform(golf['Windy'])
```

```python
le5 = LabelEncoder()
golf['Play'] = le5.fit_transform(golf['Play'])
```

```python
golf
```

```python
X = golf.iloc[:, :-1]
y = golf.iloc[:, -1]
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
model = CategoricalNB()
```

```python
model.fit(X, y)
```
